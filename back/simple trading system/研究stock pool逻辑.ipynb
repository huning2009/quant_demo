````python
def stock_pool(start, end):
    """
    Arguments
        start -- start date
        end -- end date
    
    Returns
        tuple, all adjust date & {adjust_date:code_list}
    """
    # initial the dict of adjust code of the adjust date
    adjust_date_codes_dict = dict()
    
    # all trading date
    all_dates = get_trading_dates(start, end)
    
    # initial the last phase codes pool
    # phase : 期
    last_phase_codes = []
    
    # 调仓周期
    adjust_interval = 7
    
    # store the adjust dates
    all_adjust_dates = []
    
    # the size of the stock pool
    size = 100
    
    # start:0, stop:len(all_dates), interval:adjust_interval
    # if you want the first trade date of the month
    # first use pd.date_range('2015-01-01', '2017-07-21', freq='M')
    # return the last date of the month, and use next 1 trade date function
    # next 1 trade date : all_dates[all_dates.index(last_day_of_the_month) + 1]
    for _index in range(0, len(all_dates), adjust_interval):
        # save the adjust date
        adjust_date = all_dates[_index]
        all_adjust_dates.append(adjust_date)
        
        print('调整日期： %s' % adjust_date, flush=True)
        
        # 获取(0,30)区间内pe值最小的前100的codes
        daily_cursor = daily.find(
            {'date': adjust_date, 'pe': {'$gt':0, '$lt':30},
            'index': False, 'is_trading':True},
            sort=[('pe', ASCENDING)],
            projection={'code':True, '_id':False},
            limit=100
        )
        codes = [daily['code'] for daily in daily_cursor]
        
        # initial the current phase stock pool
        this_phase_codes = []
        
        # 查询出上期股票池仍处于停牌的股票, 这类股票是不能进行交易的
        if len(last_phase_codes) > 0:
            suspension_cursor = daily.find(
                {'code': {'$in' : last_phase_codes}, 'date':adjust_date, 'is_trading' : False},
                projection = {'code':True, '_id':False}
            )
            suspension_codes = [x['code'] for x in suspension_cursor]
            
            # 将停牌的票存到本期股票池中
            this_phase_codes = suspension_codes
        
        print('上期停牌', flush=True)
        print(this_phase_codes, flush=True)
        
        # 用新的股票将剩余位置补齐, 一期的票要有多少只? 100只？
        # 假定100只, if let you do, how to do?
        need_nums = size - len(this_phase_codes)
        this_phase_codes += codes[:need_nums]
        
        last_phase_codes = this_phase_codes
        adjust_date_codes_dict[adjust_date] = this_phase_codes
        
        print('最终出票', flush=True)
        print(this_phase_codes, flush=True)
    
    return (all_adjust_dates, adjust_date_codes_dict)
````

````python
def find_out_stocks(last_phase_codes, this_phase_codes):
    """
    找出需要被卖出的股票
    
    如果要加上涨跌停限制, 只需要在判断之后, 移除不可能买入或者卖出的票
    
    停牌的票不能卖出, 那什么样的票需要被卖出呢?
    
    不是对两个列表做差集, 233
    光速打脸啊, 上期的票不在本期了, 不就是表明要被调出了吗
    
    用循环的好处之一, 不容易搞混淆,且可以保持顺序,虽然是倒序,不知道有什么意义
    貌似也没什么意义了,毕竟有停牌的存在,所以到了后面也肯定无法维持顺序的
    用set需要注意的是,不在某个集合里,就要减掉
    list(set(last_phase_codes) - set(this_phase_codes))
    """
    out_stocks = []
    for code in last_phase_codes:
        if code not in this_phase_codes:
            out_stocks.append(code)
    return out_stocks
````

````python
def evaluate_stock_pool():
    """
    对股票池做一个简单的评价,回测前的回测
    """
    start = '2015-01-01'
    end = '2015-12-31'
    adjust_dates, codes_dict = stock_pool(start, end)
    
    df_profit = pd.DataFrame(columns=['profit', 'hs300'])
    
    df_profit.loc[adjust_dates[0]] = {'profit':0, 'hs300':0}
    
    hs300_begin_value = daily.find_one({'code':'000300', 'date':adjust_dates[0], 
                                        'index':True})['close']
    # 通过净值表达累计收益
    net_value = 1
    
    for _index in range(1, len(adjust_dates) - 1):
        last_phase_date = adjust_dates[_index -1]
        current_phase_date = adjust_dates[_index]
        
        # 获取上一期股票池
        codes = codes_dict[last_phase_date]
        
        # 构建股票代码和后复权买入价格的股票?? 后复权买入??
        code_buy_close_dict = dict()
        
        buy_daily_cursor = daily_hfq.find(
            {'code':{'$in':codes}, 'date':last_phase_date},
            projection={'close':True, 'code':True, '_id':False}
        )
        
        for buy_daily in buy_daily_cursor:
            code = buy_daily['code']
            code_buy_close_dict[code] = buy_daily['close']
        
        # 上一期股票池在调整日当天的价格, 居然用收盘价...   
        sell_daily_cursor = daily_hfq.find(
            {'code':{'$in':codes}, 'date':current_phase_date},
            projection={'close':True, 'code':True, '_id':False}
        )
        
        # 计算单期收益
        profit_sum = 0
        count = 0
        for sell_daily in sell_daily_cursor:
            code = sell_daily['code']
            
            # 这一步的判断是为了确保code在dict中，不会报错
            if code in code_buy_close_dict:
                buy_close = code_buy_close_dict[code]
                sell_close = sell_daily['close']
                
                # 因为是用净值代表, 话说利润率能相加吗9 9,并且最后还拿来相除
                # 这样做也未免太粗暴了吧，每只股票价格不一样，即便是等金额分配资金
                profit_sum += (sell_close - buy_close) / buy_close
                
                count += 1
        # 有多少只股票，就要除以相应的份数    
        if count > 0:
            profit = np.round(profit_sum / count, 4)

            hs300_close = daily.find_one({'code':'000300', 'date':current_phase_date,
                                         'index':True})['close']

            # 计算净值和累计收益
            net_value *= (1 + profit)
            df_profit.loc[current_phase_date] = {
                'profit': np.round((net_value - 1) * 100, 4),
                'hs300': np.round((hs300_close - hs300_begin_value) * 100 / hs300_begin_value, 4)
            }
        
    # 绘制曲线
    df_profit.plot(title='Stock Pool Evaluation Result', kind='line');
    # show picture
#     plt.show()
````
